```python
appl_c = pd.read_csv(company_list[0])
amzn = pd.read_csv(company_list[1])
goog = pd.read_csv(company_list[2])
msft = pd.read_csv(company_list[3])
```

```python
closing_price = pd.DataFrame()
```

```python
closing_price['AAPL'] = appl_c['close']
closing_price['AMZN'] = amzn['close']
closing_price['GOOG'] = goog['close']
closing_price['MSFT'] = msft['close']
```

```python
closing_price
```

```python
sns.pairplot(closing_price)
```

```python
closing_price.corr()
```

```python
sns.heatmap(closing_price.corr() ,annot = True)
```

```python
closing_price
```

```python
(closing_price['AAPL'] - closing_price['AAPL'].shift(1)) / closing_price['AAPL'].shift(1)*100
```

```python
for col in closing_price.columns:
 closing_price[col + '_pct_change'] =  closing_price[col] = (closing_price[col]-closing_price[col].shift(1)) / closing_price[col].shift(1) * 100
```

```python
closing_price.columns
```

```python
closing_p = closing_price[['AAPL_pct_change','AMZN_pct_change','GOOG_pct_change','MSFT_pct_change']]
```

```python
closing_p
```

```python
h = sns.PairGrid(closing_p)
h.map_diag(sns.histplot)
h.map_lower(sns.scatterplot)
h.map_upper(sns.kdeplot)
```
