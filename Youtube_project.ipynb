```python
TextBlob("Logan Paul it's yo big day ‚ÄºÔ∏è‚ÄºÔ∏è‚ÄºÔ∏è	").sentiment.polarity
```

```python
polarity =[]
for comment in df['comment_text']:
#Indent the try-except block within the for loop
 try :

  polarity.append(TextBlob(comment).sentiment.polarity)
 except:

    polarity.append(0)
```

```python
df['polarity'] = polarity
```

```python
df.head(5)
```

```python
filtter1 = df['polarity']==1
```

```python
positive_comments = df[filtter1]
```

```python
filtter2 = df['polarity']==-1
```

```python
negative_comments = df[filtter2]
```

```python
def punc_count(text):
 return len([char for char in text if char in string.punctuation])
```

```python
sample_csv = full_df_csv[0:10000]
```

```python
sample_csv['count_punc'] = sample_csv['title'].apply(punc_count)
```

```python
sample_csv['count_punc']
```
